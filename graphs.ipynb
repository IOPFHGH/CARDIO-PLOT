```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/cardiovascular-disease-dataset/cardio_train.csv", sep=";")
df
```

```python
years= (df['age'] / 365).round().astype('int')
df.insert(3, "years",years)
```

```python
sns.countplot(x='years', hue='cardio', data = df,palette="Set2");
```

```python
df_cate = df.loc[:,['cholesterol','gluc', 'smoke', 'alco', 'active']]
sns.countplot(x="variable", hue="value",data= pd.melt(df_cate));
```
